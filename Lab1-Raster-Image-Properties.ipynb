# Buffer and Channel

```python
#Buffe and Channel  Sample

sample[50:150 , : , 0] = 255 # full intensity to those pixel's R channel
plt.figure( figsize = (5,5))
plt.imshow(sample)
plt.show()

sample[200:300 , : , 1] = 255 # full intensity to those pixel's G channel
plt.figure( figsize = (5,5))
plt.imshow(sample)
plt.show()

sample[350:450 , : , 2] = 255 # full intensity to those pixel's B channel
plt.figure( figsize = (5,5))
plt.imshow(sample)
plt.show()

sample[ 50:450 , 400:600 , [0,1,2] ] = 200  # adding a grayscal in the middle
plt.figure( figsize = (5,5))
plt.imshow(sample)
plt.show()
```

# Image Levels : Gray Scale and Negative

```python
''' 
    Gray Scale
    
    Y = 0.299 R + 0.587 G + 0.114 B 
'''
#Formula ::: https://en.wikipedia.org/wiki/Grayscale#Converting_color_to_grayscale
```

```python
gray = lambda rgb : np.dot(rgb[... , :3] , [0.299 , 0.587, 0.114]) 
gray = gray(sample)  


plt.figure( figsize = (5,5))
plt.imshow(gray, cmap = plt.get_cmap(name = 'gray'))
plt.show()
```

```python
#Negative

'''
log transform
-> s = c*log(1+r)

So, we calculate constant c to estimate s
-> c = (L-1)/log(1+|I_max|)

'''

max_=np.max(gray)

def log_transform():
    return(255/np.log(1+max_))*np.log(1+gray)

plt.figure(figsize=(5,5))
plt.imshow(log_transform(), cmap=plt.get_cmap(name='gray'))
plt.axis('off');
```
